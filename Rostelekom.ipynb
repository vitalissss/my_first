```python
def permutation(word1, first_char=None):
    if word1 == None or len(word1) == 0: return []
    if len(word1) == 1: return [word1]

    result = []
    first_char = word1[0]
    print(first_char)
    for sub_word in permutation(word1[1:], first_char):
        result += insert (first_char, sub_word)
    return sorted(result)
```

```python
def insert(ch, sub_word):
    arr = [ch + sub_word]
    for i in range(len(sub_word)):
        arr.append(sub_word[i:] + ch + sub_word[:i])
    return arr
```
